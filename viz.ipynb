```python
from datasets import load_dataset

dataset = load_dataset("mikronai/svg-svgrepo")
```

```python
train = dataset["train"]
```

```python
import matplotlib.pyplot as plt
import numpy as np
import random
from svgai.img import center_pad_image
from svgai.svg import render_fit

images = []
for i in range(100):
    svg = train[random.randint(0, len(train) - 1)]["item_svg"]
    img = render_fit(svg, width=64, height=64, background="white")
    img = center_pad_image(img, 64, 64)
    images.append(img)

# Create the 10x10 subplot grid
fig, axes = plt.subplots(10, 10, figsize=(20, 20))

# Flatten axes array for easier indexing
axes_flat = axes.flatten()

# Plot each image with a title
for i in range(100):
    ax = axes_flat[i]

    # Display the image
    im = ax.imshow(images[i])

    # Set title for each subplot
    ax.set_title(, fontsize=8, pad=2)

    # Remove axis ticks for cleaner look
    ax.set_xticks([])
    ax.set_yticks([])

# Adjust spacing between subplots
plt.tight_layout()
plt.subplots_adjust(top=0.95, hspace=0.3, wspace=0.1)

# Show the plot
plt.show()
```
